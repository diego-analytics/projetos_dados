## Utilizando modelos para prever a rotatividade

Nesta etapa vamos preparar os dados e utilizá-los para criar dois modelos de previsão da rotatividade dos clientes: um de regressão logística e outro de floresta aleatória. Depois de testá-los, vamos avaliá-los quanto a acurácia, precisão e sensibilidade.

Para iniciar, vamos criar duas variáveis, X e y, com as características e a variável objetivo, respectivamente. Em seguida, vamos dividir os dados em um conjunto de treinamento e um conjunto de teste, na proporção de 80/20. Para garantir maior regularidade aos dados, vamos também normalizá-los, ou seja, transformaremos suas distribuição para que ela tenha média 0 e desvio padrão 1. Isso pode não alterar significativamente os dados de algumas características, mas com certeza garante um funcionamento mais regular dos nossos modelos.

```python
# Criando a "y" contendo a variável objetivo, ou seja, "churn"
y = gym_data['churn']

# Criando "X" contendo as características
X = gym_data.drop('churn', axis=1)

# Dividindo o conjunto de dados em dados de treinando e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Criando um objeto StandardScaler para normalizar os dados e aplicando-o ao conjunto de treino e treiná-lo ao mesmo tempo
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)

# Transformando os dados do conjunto de teste
X_test_st = scaler.transform(X_test)
```

Agora vamos usar nossos dados para treinar um modelo de regressão logística e realizar algumas previsões, que salvaremos na variável ``log_prev``. Por fim, vamos testar o modelo quanto a acurácia, sensibilidade e precisão.

```python
# Criando o modelo do tipo LogisticRegression
log_model = LogisticRegression(max_iter=500)

# Treinando o modelo
log_model.fit(X_train_st, y_train)

# Fazendo previsões
log_prev = log_model.predict(X_test_st)

# Testando o modelo
print('EAM:{:.2f} EQM:{:.2f} R2:{:.2f} '.format(mean_absolute_error(y_test, log_prev), 
                                          mean_squared_error(y_test, log_prev), r2_score(y_test, log_prev)))
```

Vamos fazer o mesmo agora para um modelo de floresta aleatória.

```python
# Criando o modelo do tipo RandomForestRegressor
rf_model = RandomForestRegressor()

# Treinando o modelo
rf_model.fit(X_train_st, y_train)

# Fazendo previsões
rf_prev = rf_model.predict(X_test_st)

# Testando o modelo
print('EAM:{:.2f} EQM:{:.2f} R2:{:.2f} '.format(mean_absolute_error(y_test, rf_prev), 
                                          mean_squared_error(y_test, rf_prev), r2_score(y_test, rf_prev)))
```

Comparando os dois modelos, temos que a regressão logística tem um Erro Absoluto Médio (EAM) menor, porém o modelo se mostra inferior à floresta aleatória no Erro Quadrático Médio (EQM) e no coeficiente R2. EAM mede a acurácia total do modelo (é o somatório do módulo de cada erro da previsão), já EQM eleva os valores de cada erro ao quadrado antes de somá-los, o que o torna mais sensível a erros grandes, medindo assim a sensibilidade do modelo. Comparando essas duas métricas, a regressão logística parece ser mais acertiva nas previsões, mas, em média, têm erros maiores. Já o coeficiente R2 mede quanto o modelo foi capaz de prever da variável objetivo, ou seja, quão preciso foi seu desempenho. Ela indica que o modelo da floresta aleatória conseguiu prever uma parcela total maior da variável objetivo.

Com essas informações, podemos considerar que o modelo **floresta aleatória** é superior ao modelo de regressão logística para prever a rotatividade dos clientes da rede Model Fitness.

## Agrupando os clientes

Vamos estudar os dados e utilizar aprendizado de máquina para dividir os usuários em grupos de acordo com suas características. Isso pode ser útil para fins de marketing e criação de promoções diferenciadas para cada grupo. Para começar precisamos padronizar os dados, de forma similar ao que fizemos com os dados de treino e teste.

```python
# Criando um objeto do tipo "StandardScaler" para utilizarmos na padronização
scaler2 = StandardScaler()

# Padronizando nossa varíavel X
x_sc = scaler2.fit_transform(gym_data.drop(columns = ['churn']))
```

Agora vamos usar a função ``linkage()`` para criar uma matriz de distâncias e, a partir dela, criaremos um dendrograma. Isso nos permitirá saber o número ideal de grupos em que os clientes devem ser divididos.

```python
# Criando a matriz de distâncias a partir dos dados padronizados
linked = linkage(x_sc, method='ward')

# Criando o dendograma para a matriz
plt.figure(figsize=(15,8))
dendrogram(linked, orientation='top')
plt.title('Agrupamento hierárquico para os dados dos usuários da rede Model Fitness')
plt.show()
```

Pelo dendrograma, podemos distinguir 5 grandes grupos. Vamos usar esse número para agrupar nossos usuários, através do algoritmo *K-Means*.

```python
# Definindo o número de agrupamentos
km = KMeans(n_clusters=5)

# Aplicando o algoritmo aos dados para formar o vetor de agrupamento
labels = km.fit_predict(x_sc)
```

Na variável ``labels`` nós temos o índice dado a cada usuário pelo algoritmo. Essas valores vão de 1 a 5, sem importância da ordem. Vamos adicionar essa informação ao nosso DataFrame, criando a coluna ``km_cluster``.

```python
# Adicionado a coluna
gym_data['km_cluster'] = labels

# Exibindo como ficaram as primeiras linhas da tabela
gym_data.head()
```

Vamos agrupar os dados por grupo e ver como diferem as médias:

```python
gym_data.groupby('km_cluster').mean()
```

Vemos que existem médias de alguns grupos com valores bem marcantes. O grupo 0 é formato totalmente por pessoas que não informaram um telefone para contato. Já o grupo 1 tem usuários cujo período do contrato é bastante alto em relação aos outros. O grupo 2 parece combinar pessoas que não são funcionários de empresas parceiras e não foram indicados por amigos. O grupo 3 parece ser totalmente formado por pessoas que não moram ou trabalham próximo da academia. E o grupo 4 reune pessoas que moram/trabalham próximo, gastam pouco em despesas adicionais e tem a menor média de meses até o fim do contrato.

## Conclusões

Observamos, nas análises realizadas, que algumas características parecem ter um impacto maior na rotatividade dos usuários. Pessoas que moram ou trabalham próximo a academia, quem conseguiu um desconto por ser indicado por um amigo, funcionários de companhias parceiras e clientes que contrataram pacotes de maior duração tendem a saírem menos. * Buscar novas empresas parceiras, dado que esse tipo de usuário tende a continuar;
* Tomar medidas para popularizar o desconto por indicação e incentivar os frequentadores atuais a indicarem amigos;
* Criar promoções especiais para clientes que se aproximem de uma frequência semanal menor que 2;
* Oferecer descontos ou brindes para os contratantes de pacotes maiores.
